```python
g = np.array(['rock', 'paper', 'scissors'])
g
```

```python
np.random.choice(g)
```

```python
for _ in range(10):
    p1 = np.random.choice(g)
    p2 = np.random.choice(g)
    print(f'{p1}-{p2}')
```

```python
pet = np.array(['cat', 'dog', 'bird', 'fish'])
c = np.random.choice(pet, size=100, replace=True)
c
```

```python
u, count = np.unique(c, return_counts=True)
print(u)
print(count)
```

```python
plt.bar(u, count)
```

```python
pet = np.array(['cat', 'dog', 'bird', 'fish'])
c = np.random.choice(pet, size=100, replace=True, p=[.2, .4, .1, .3])
c
```

```python
u, count = np.unique(c, return_counts=True)
print(u)
print(count)
plt.bar(u, count)
```
